```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("student_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
 df.select_dtypes(include=['object']).columns
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop("G3", axis=1)
Y = df["G3"]
```

```python
X = df.drop(["G1", "G2", "G3"], axis=1)
Y = df["G3"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error
Y_pred = model.predict(X_test)
print("MAE:", mean_absolute_error(Y_test, Y_pred))
```

```python
 import seaborn as sns
 sns.scatterplot(x=Y_test, y=Y_pred)
 plt.xlabel("Actual")
 plt.ylabel("Predicted")
 plt.plot([0,20], [0,20], color='red')
 plt.show()
```

1. The model shows a positive correlation between actual and predicted scores.
2. Most predictions follow the general trend, indicating the model is learning patterns.
3. Some deviations exist, suggesting the model can be improved further.
